#### **Задание 1** 

Напишите функцию, которая бы по данным о температуре для конкретного блока (тип **Series**) определяла области, в которых происходит охлаждение блока. Параметр **threshold** задает минимальную длительность такой области.

```python
def find_decreasing_regions(series, threshold = 10):
    diff = series.diff()
    diff.loc[lambda x : x >= 0] = 1 # Mark increasing regions with 1
    diff.loc[lambda x : x <  0] = 0 # Mark decreasing regions with 0
    posedge_negedge = diff.diff()
    start = posedge_negedge[posedge_negedge==-1].index - 1
    end   = posedge_negedge[posedge_negedge== 1].index - 2


    # align start and end
    if len(start) > len(end):
        end   = end  .to_series(index=np.arange(0, len(start)-1))
        start = start.to_series(index=np.arange(0, len(start)))
    elif len(end) > len(start):
        end   = end  .to_series(index=np.arange(0, len(end)))
        start = start.to_series(index=np.arange(1, len(end)))
    elif end[0] < start[0]:
        end   = end  .to_series(index=np.arange(0, len(end)))
        start = start.to_series(index=np.arange(1, len(end)+1))


    res = pd.DataFrame(columns=['start', 'end'], data=np.empty(shape=[max(len(start), len(end)), 2], dtype=int))
    res['start'] = start
    res['end']   = end
    res = res.loc[(res['end'] - res['start']) >= 10]
    res = res.astype(int)
    return res
```

Определим области охлаждения для каждого блока

```python
block_names = ['temp1','temp2','temp3','temp4','temp5','temp6','temp7','temp8','temp9']
decreasing_regions = [find_decreasing_regions(df[name]) for name in block_names]
```

```python
plt.figure(figsize=(15,12))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.plot(df['temp'+str(i+1)])
    xlim = 200
    plt.xlim(0,xlim)
    for index, row in decreasing_regions[i].iterrows():
        if row['end'] > xlim:
            break
        plt.axvspan(row['start'], row['end'], color='red', alpha=0.5)
    plt.title('Block '+str(i+1))
plt.show()
```

#### **Задание 4** 

Напишите функцию, которая бы по известной текущей температуре **temp**, верхнего **max_temp** и нижнего **min_temp** температурных порогов определяла состояние блока в диапазоне от 0 до 1 (0 - блок имеет нижнюю температуру, 1 - верхнюю). При этом значения температуры, выше верхнего порога должны соответствовать состоянию 0, а температуры, ниже нижнего порога - состоянию 1. 

```python
def get_state(temp, max_temp, min_temp):

    mms = lambda temp, max_temp, min_temp: (temp - max_temp) / (min_temp - max_temp)

    res = None

    if temp > max_temp:
        res = 0
    elif temp < min_temp:
        res = 1
    else:
        res = mms(temp=temp, max_temp=max_temp, min_temp=min_temp)

    return res
```
